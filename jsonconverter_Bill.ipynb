```python
data = json.loads(json_data)
bills = data['QueryResponse']['Bill']
```

```python
rows = []

today = datetime.now(tz=None)

for bill in bills:
    invoice_id = bill['Id']
    create_time = bill['MetaData']['CreateTime']
    due_date = bill['DueDate']
    customer_name = bill['VendorRef']['name']
    balance = bill['Balance']

    create_time = pd.to_datetime(create_time).tz_localize(None)
    due_date = pd.to_datetime(due_date).tz_localize(None)

    time_remaining = due_date - today

    rows.append([invoice_id, create_time, customer_name, balance, due_date, time_remaining])

df = pd.DataFrame(rows, columns=['Id', 'CreateTime', 'CustomerName', 'Balance', 'Due Date', 'Remaining Time'])
df
```
